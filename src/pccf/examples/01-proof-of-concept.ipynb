Input signal is 

- $X[i] \sim N(\mu_0, \sigma)$ before change
- $X[i] \sim N(\mu_0 + \delta, \sigma)$ after change

CuSum output statistic is calculated as 

- $S[i] = X[i] - \mu_0$ if $i == 0$
- $S[i] = S[i-1] + X[i] - \mu_0$

- If $X[i] \sim N(\mu_0, \sigma)$ then $X[i] - \mu_0 \sim N(0.0, \sigma)$ before change, 
  - and therefore $S[i] = S[i-1] + X[i] - \mu_0$ is a random walk $S[i] = \sum_{i=0}^{k} (x_i \sim N(0.0, \sigma)) $

- After the change $X[i] - \mu_0 \sim N(\delta, \sigma)$ and therefore $S[i]$ will be a random walk with the drift.
